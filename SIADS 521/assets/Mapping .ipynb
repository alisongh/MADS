```python
## Import folium
# What is folium? Folium is a powerful Python library that helps you create several types of 
# Leaflet maps. The fact that the Folium results are interactive makes this library very useful 
#for dashboard building. 
import folium
```

https://blog.dominodatalab.com/creating-interactive-crime-maps-with-folium/

```python
## Show San Francisco on the global map 
SF_center = (37.774929, -122.419416)
SF = folium.Map(location=SF_center, zoom_start=10)
SF
```

```python
## If we want to show where the restuarant is at in the world map 
## Change the style of the map
Tira_kit = (res['business_latitude'][0],res['business_longitude'][0])
Tira_kit = folium.Map(location=Tira_kit, zoom_start=30, tiles='Stamen Toner')
Tira_kit
```

```python
## Add a marker
Tira_kit = (res['business_latitude'][0],res['business_longitude'][0])
folium.Marker(location=Tira_kit,popup='Tiramisu Kitchen').add_to(SF)
SF
```

```python
## Add a different type of marker-- circle(has to define radius)
j_donut = (res['business_latitude'][16],res['business_longitude'][16])
folium.CircleMarker(
    location=j_donut,
    radius=10,
    popup='Jelly Donut',
    color='yellow',
    fill=True,
    fill_color='yellow',
).add_to(SF)
SF
```

```python
## Drop all dulipcates inspection_score
res["risk_category"]= res["risk_category"].replace("Moderate Risk",1) 
res["risk_category"]= res["risk_category"].replace("Low Risk",0) 
res["risk_category"]= res["risk_category"].replace("High Risk",2) 
group=res.set_index('business_name')
group= group.drop_duplicates('inspection_date')
group.head(10)
```

```python
group1= group[['inspection_score', 'risk_category', 'business_id']].groupby('business_id')
group1.head(10)
```

```python
stats = group1.agg('mean').sort_values(by='inspection_score', ascending=False)
stats.head(20)
```

```python
counts = group1.size()
counts.head(10)
```

```python
stats = stats.assign(count=counts)
stats.head(200)
```

```python
from branca.colormap import linear

### OrRd colormap is kind of similar to tabelog score colors
colormap = linear.OrRd_03.scale(
    stats['inspection_score'].min(),
    stats['inspection_score'].max())

colors = stats['inspection_score'].apply(colormap)
colors
```

```python
state_geo = gpd.read_file('assets/us-states.json')

# Load the unemployment value of each state
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data
crime_data = pd.read_csv('assets/stateCrime.csv')

# Initialize the map:
m = folium.Map(location=[37, -102], zoom_start=4)
 
# Add the color for the chloropleth:
m.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=crime_data,
 columns=['State', 'Murder.Rate'],
 key_on='feature.properties.name',
 fill_color='RdBu',
 fill_opacity=0.8,
 line_opacity=0.3,
 legend_name='Crime Rate (%)'
)
folium.LayerControl().add_to(m)
 
m
```

```python
# Custom it with the same argument as 1D density plot
sns.kdeplot(happiness['Happiness.Score'], happiness['Freedom'], cmap="Reds", shade=True, bw=.15)
```

```python
# Some features are characteristic of 2D: color palette and wether or not color the lowest range
sns.kdeplot(happiness['Happiness.Score'], happiness['Freedom'],happiness['Generosity'], cmap="Blues", shade=True, shade_lowest=True, )
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

w = 20
h = 20
d = 50
plt.figure(figsize=(w, h), dpi=d)

fig = plt.figure()
ax = fig.add_subplot(111)
x = happiness['Happiness.Score']
y = happiness['Freedom']
x, y = np.meshgrid(x, y)
z = np.sin(x * np.pi / 2) + np.cos(y * np.pi / 3)

plt.contour(x, y, z)
```
